```python
#img_path = 'avokado.jpg'
# Första bilden togs på avokado som låg på bordet med en prickig duk, prediktionen sa att det kan vara en duschdraperi.
# Testade med en ny bild genom att använda en annan bakgrund
img_path = 'avokado2.jpg'
img = image.load_img(img_path, target_size=(224, 224))
img
```

```python
x = image.img_to_array(img)
x.shape
```

```python
x = np.expand_dims(x, axis=0)
x.shape
```

```python
x = preprocess_input(x)
```

```python
preds = model.predict(x)

print('Predicted:', decode_predictions(preds, top=3)[0])
```

# 2. Answer the question: Why is "Pretrained models" a very powerful concept?

1. Man besparar ganska mycket tid genom att använda pretrained modeller, eftersom man inte behöver skapa sin egen modell från grunden som tar mycket tid.

2. Pretrained modeller är tränade på stora datasets som är också en enorm fördel. Skapar man sin egen modell, då kan det vara utmanande att hitta lika stora mängder av representativ data.

3. De är testade och validerade som gör de mycket pålitliga.
